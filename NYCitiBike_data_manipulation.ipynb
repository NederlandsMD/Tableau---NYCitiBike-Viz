```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
Trips_Jan = pd.read_csv("Trips_2018_01.csv")
Trips_Jan.head()
```

```python
Trips_Jun = pd.read_csv("Trips_2018_06.csv")
Trips_Jun.head()
```

```python
# Concatenate the dataframes so we have one dataset

frames = [Trips_Jan, Trips_Jun]
Summer_Winter_df = pd.concat(frames)
Summer_Winter_df.reset_index(inplace=True, drop=True)
```

```python
Summer_Winter_df.iloc[1767568,1]
Summer_Winter_df.to_csv("Summer_Winter_df.csv")
```

```python
# Basic Checks on the merged dataset
Summer_Winter_df.iloc[1200000:1200010]
len(Summer_Winter_df)
Summer_Winter_df.describe()
```

```python
#Create a variable for distance biked

Trip_distance = []

for x in range(len(Summer_Winter_df)):
    trip_hrs = Summer_Winter_df.iloc[x,0]/3600
    if (trip_hrs <= 2):
        Trip_distance.append(trip_hrs * 7.456)
    else:
        Trip_distance.append(14.9)

print(len(Trip_distance))
```

```python
# Add distance biked to the DF, then group all by bikeid (in the month of June... our most recent month)
Summer_Winter_df["Distance"] = Trip_distance

June_df = Summer_Winter_df.iloc[718994:-1,:]

bikes = June_df.groupby(["bikeid"])
bikes = bikes["Distance"].sum()
bikes = pd.DataFrame(bikes)
bikes = bikes.sort_values("Distance", ascending=False)
bikes.head()
bikes.to_csv("Citi_Bikes.csv")
```

```python
# Create a DF of distinct bike stations, where the columns are Popularity Start Station (Summer),
# End Station (Summer), Start Station (Winter), and End Station (Winter). 
# We will evenutally need id, name, latitude, and longitude here, then the four columns we've discussed

Station_df = pd.DataFrame(columns=['Station_ID','Station_name','Lat', 'Lng', 'Summer_Pop_Start', 'Summer_Pop_End', 
                                   'Summer_Dist', 'Winter_Pop_Start', 'Winter_Pop_End', 'Winter Dist'])

Start_Station_array = Summer_Winter_df["start station id"].unique()

Station_array = Start_Station_array.tolist()

#  Key columns... for start station (ID-3, Name-4, Lat-5, Lng-6), for end station (ID-7, Name-8, Lat-9, Lng-10)
for i in range(len(Summer_Winter_df)):
    EndStation = Summer_Winter_df.iloc[i, 7]
    if EndStation not in Station_array:
        Station_array.append(EndStation)

Start_Station_array = Start_Station_array.tolist()

Station_df["Station_ID"] = Station_array

for station in Station_df["Station_ID"]:
    if station in Start_Station_array: 
        info_row = Summer_Winter_df[Summer_Winter_df["start station id"]==station].iloc[0,:]
        Station_df.loc[Station_df["Station_ID"]==station, "Station_name"] = info_row[4]
        Station_df.loc[Station_df["Station_ID"]==station, "Lat"] = info_row[5]
        Station_df.loc[Station_df["Station_ID"]==station, "Lng"] = info_row[6]
    else:
        info_row = Summer_Winter_df[Summer_Winter_df["end station id"]==station].iloc[0,:]
        Station_df.loc[Station_df["Station_ID"]==station, "Station_name"] = info_row[8]
        Station_df.loc[Station_df["Station_ID"]==station, "Lat"] = info_row[9]
        Station_df.loc[Station_df["Station_ID"]==station, "Lng"] = info_row[10]
    
Station_df.head()

        
```

```python
# Now that we have the dataframe set up with Station ID, Name, Lat and Lng, we need to do some counts

# IDEA: create a sliced dataframe with iloc="Station_ID", then take len of the sliced df and drop in.

for my_station in Station_df['Station_ID']:
    try:
        temp_df_WS = Trips_Jan.loc[Trips_Jan['start station id']==my_station,:]
        Station_df.loc[Station_df['Station_ID']==my_station,"Winter_Pop_Start"] = len(temp_df_WS)
    except: 
        Station_df.loc[Station_df['Station_ID']==my_station,"Winter_Pop_Start"] = 0
    try:
        temp_df_WE = Trips_Jan.loc[Trips_Jan['end station id']==my_station,:]
        Station_df.loc[Station_df['Station_ID']==my_station,"Winter_Pop_End"] = len(temp_df_WE)
    except:
        Station_df.loc[Station_df['Station_ID']==my_station,"Winter_Pop_End"] = 0
    try:
        temp_df_SS = Trips_Jun.loc[Trips_Jun['start station id']==my_station,:]
        Station_df.loc[Station_df['Station_ID']==my_station,"Summer_Pop_Start"] = len(temp_df_SS)
    except:
        Station_df.loc[Station_df['Station_ID']==my_station,"Summer_Pop_Start"] = 0
    try:
        temp_df_SE = Trips_Jun.loc[Trips_Jun['end station id']==my_station,:]
        Station_df.loc[Station_df['Station_ID']==my_station,"Summer_Pop_End"] = len(temp_df_SE)
    except:
        Station_df.loc[Station_df['Station_ID']==my_station,"Summer_Pop_End"] = 0
        
Station_df.head()

```

```python
# Set up Trips Jan and Trips June with Trip Duration variables in minutes, then filter

# Trips_Jan.head()
# Trips_Jan["TD_Mins"] = Trips_Jan["tripduration"] / 60
# Trips_Jan.head()
print(f"Trips Jan length: {len(Trips_Jan)}")

# I AM FILTERING OUT TRIPS OVER 6 HOURS, I THINK THEY MESS WITH THE DATA AND COULD BE FLAWED (cuts 371 trips)

# Trips_Jan_filtered = Trips_Jan.loc[Trips_Jan['TD_Mins'] <= 360, :]
print(f"Trips Jan filtered length: {len(Trips_Jan_filtered)}")

# Trips_Jun["TD_Mins"] = Trips_Jun["tripduration"] / 60
print(f"Trips Jun length: {len(Trips_Jun)}")

# I AM FILTERING OUT TRIPS OVER 6 HOURS, I THINK THEY MESS WITH THE DATA AND COULD BE FLAWED (cuts 714 trips)

Trips_Jun_filtered = Trips_Jun.loc[Trips_Jun['TD_Mins'] <= 360, :]
print(f"Trips Jun filtered length: {len(Trips_Jun_filtered)}")
```

```python
# Group Trips_Jan_filtered and Trips_Jun_filtered by start station id
station_group_winter = Trips_Jan_filtered.groupby(['start station id'])
station_group_winter = station_group_winter["TD_Mins"].mean()
station_group_winter = pd.DataFrame(station_group_winter)
station_group_winter.reset_index(drop=False, inplace=True)
station_group_winter.head()
print(f"Stations in Winter: {len(station_group_winter)}")

station_group_summer = Trips_Jun_filtered.groupby(['start station id'])
station_group_summer = station_group_summer["TD_Mins"].mean()
station_group_summer = pd.DataFrame(station_group_summer)
station_group_summer.reset_index(drop=False, inplace=True)
station_group_summer.head()
print(f"Stations in Summer: {len(station_group_summer)}")

```

```python
# Now we need to read in the correct values into our stations dataframe

# my_station = 315
# winter_dist = station_group_winter[station_group_winter['start station id']==my_station].iloc[0,1]
# print(f"Winter Avg Distance, station 315: {winter_dist}")
# Station_df.loc[Station_df['Station_ID']==my_station,"Winter Dist"] = winter_dist
# Station_df.head()

for my_station in Station_df['Station_ID']:
    try:
        winter_dist = station_group_winter[station_group_winter['start station id']==my_station].iloc[0,1]
        Station_df.loc[Station_df['Station_ID']==my_station,"Winter Dist"] = winter_dist
    except:
        Station_df.loc[Station_df['Station_ID']==my_station,"Winter Dist"] = 'NaN'
    try:
        summer_dist = station_group_summer[station_group_summer['start station id']==my_station].iloc[0,1]
        Station_df.loc[Station_df['Station_ID']==my_station,"Summer_Dist"] = summer_dist
    except:
        Station_df.loc[Station_df['Station_ID']==my_station,"Summer_Dist"] = 'NaN'
        
Station_df.head(20)
```
